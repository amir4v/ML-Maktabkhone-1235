```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('./Iris.csv')

X = df.iloc[:, 1:4].values
y = df.iloc[:, 4].values

df.head()
```

```python
df.info()
```

```python
pca = PCA(n_components=2)

X = pca.fit_transform(X)
```

```python
X
```

```python
main_df = pd.DataFrame(data=X, columns=['principal component 1', 'principal component 2'])

main_df
```

```python
pca.components_
```

```python
pca.explained_variance_
```

```python
final_df = pd.concat(
    [
        main_df,
        df[['Species']]
    ],
    axis=1
)
```
