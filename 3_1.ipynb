```python
def sinplot():
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(
            x, np.sin(x + i * 0.5) * (7 - i)
        )
```

```python
g = sns.FacetGrid(tips,
    col='time',
    row='sex',
)

g.map_dataframe(
    sns.histplot,
    'total_bill',
    'tip',
    binwidth=2
)

g.set_axis_labels('Total bill', 'Count')
```

```python
g = sns.FacetGrid(tips,
    col='time',
    
    hue='sex',
)

g.map_dataframe(
    sns.histplot,
    'total_bill',
    'tip',
    binwidth=2
)

g.set_axis_labels('Total bill', 'Count')

g.add_legend()
```
